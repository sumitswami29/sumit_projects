```python
# Assuming the DataFrame is loaded as 'ipl'

# Filter players who were sold in 2021 and their respective teams
sold_in_2021 = ipl[ipl['2021 Squad'].notna()][['Player', '2021 Squad']]

# Filter players who were unsold in 2022
unsold_in_2022 = ipl[ipl['Team'] == 'Unsold']['Player']

# Merge the data to get the team information for the unsold players
unsold_players_with_2021_team = sold_in_2021[sold_in_2021['Player'].isin(unsold_in_2022)]

# Print the names and teams of players who were sold in 2021 but unsold in 2022
for _, row in unsold_players_with_2021_team.iterrows():
    print(f"Player: {row['Player']}, Team in 2021: {row['2021 Squad']}")
```

```python
spent_players = ipl[ipl['Team'] != 'Unsold']

# Calculate the total spend by each team
total_spend_by_team = spent_players.groupby('Team')['COST IN ₹ (CR.)'].sum().sort_values(ascending=False)
print("Total Spend by Team:")
print(total_spend_by_team)

# Plotting total spend by team
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x=total_spend_by_team.index, y=total_spend_by_team.values)
plt.title('Total Spend by Team')
plt.xlabel('Team')
plt.ylabel('Total Spend in ₹ (CR.)')
plt.xticks(rotation=90)
plt.show()

# 2. Average Cost per Player by Team
# Calculate the average cost per player by each team
average_cost_per_player_by_team = spent_players.groupby('Team')['COST IN ₹ (CR.)'].mean().sort_values(ascending=False)
print("Average Cost per Player by Team:")
print(average_cost_per_player_by_team)

# Plotting average cost per player by team
plt.figure(figsize=(12, 6))
sns.barplot(x=average_cost_per_player_by_team.index, y=average_cost_per_player_by_team.values)
plt.title('Average Cost per Player by Team')
plt.xlabel('Team')
plt.ylabel('Average Cost in ₹ (CR.)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Assuming the DataFrame is loaded as 'ipl'

# Compute the correlation matrix
correlation_matrix = ipl[['COST IN ₹ (CR.)', 'Cost IN $ (000)']].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Box plot of player costs across different teams
plt.figure(figsize=(14, 8))
sns.boxplot(x='Team', y='COST IN ₹ (CR.)', data=ipl[ipl['Team'] != 'Unsold'])
plt.title('Distribution of Player Costs by Team')
plt.xlabel('Team')
plt.ylabel('Cost in ₹ (CR.)')
plt.xticks(rotation=90)
plt.show()

# Box plot of player costs across different player types
plt.figure(figsize=(12, 6))
sns.boxplot(x='TYPE', y='COST IN ₹ (CR.)', data=ipl[ipl['COST IN ₹ (CR.)'].notna()])
plt.title('Distribution of Player Costs by Player Type')
plt.xlabel('Player Type')
plt.ylabel('Cost in ₹ (CR.)')
plt.show()
```

```python
# Plot trend lines for player costs by team
plt.figure(figsize=(14, 8))
sns.lineplot(x='Team', y='COST IN ₹ (CR.)', data=ipl[ipl['Team'] != 'Unsold'], estimator='mean', marker='o')
plt.title('Trend of Average Player Costs by Team')
plt.xlabel('Team')
plt.ylabel('Average Cost in ₹ (CR.)')
plt.xticks(rotation=90)
plt.show()
```

```python


# Assuming the DataFrame is loaded as 'ipl'

# Identify players who have moved teams
# Filter out rows where the player was not in the 2021 squad
players_moved = ipl[ipl['2021 Squad'].notna() & (ipl['Team'] != 'Unsold')]

# Create a DataFrame for movements
player_movements = players_moved[['Player', '2021 Squad', 'Team']]

# Find players who moved between teams
player_movements['Movement'] = player_movements.apply(lambda row: f"{row['2021 Squad']} -> {row['Team']}", axis=1)

# Count how many times each movement occurred
movement_counts = player_movements['Movement'].value_counts()

# Display the movements
print("Player Movements:")
print(movement_counts)

# Plot the movement trends
plt.figure(figsize=(14, 8))
sns.countplot(y='Movement', data=player_movements, order=movement_counts.index)
plt.title('Player Movements Between Teams')
plt.xlabel('Count')
plt.ylabel('Movement')
plt.show()
```
